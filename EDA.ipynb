1. Hiểu Tổng Quan Về Dữ Liệu
1.1. Đọc và Xem Xét Dữ Liệu

```python
import pandas as pd
data = pd.read_csv('Amazon Customer Behavior Survey.csv')
```

```python

print(data.info())
```

```python

print(data.describe())
```

2. Kiểm tra các giá trị null

```python
print(data.isnull().sum())
```

2. Phân Tích Các Biến
2.1. Biến Định Lượng (Numerical Variables)
Phân phối của các biến định lượng:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data['age'], bins=30, kde=True)
plt.show()
```

Hộp biểu đồ (Box Plot) để phát hiện các giá trị ngoại lai:

```python
sns.boxplot(x=data['age'])
plt.show()
```

2.2. Phân Tích Biến Định Tính
Giới tính (Gender), Tần suất mua hàng (Purchase_Frequency), Danh mục mua hàng (Purchase_Categories): Sử dụng count plot để kiểm tra phân phối của các biến định tính.

```python
# Giới tính
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.show()

# Tần suất mua hàng
sns.countplot(x='Purchase_Frequency', data=data)
plt.title('Purchase Frequency Distribution')
plt.show()

# Danh mục mua hàng
sns.countplot(x='Purchase_Categories', data=data)
plt.title('Purchase Categories Distribution')
plt.show()
```

4. Phân Tích Quan Hệ Giữa Các Biến
4.1. Quan Hệ Giữa Biến Định Lượng
Tuổi và Số tiền chi tiêu: Sử dụng biểu đồ scatter để kiểm tra mối quan hệ giữa tuổi và các biến khác nếu có.

```python
sns.scatterplot(x='age', y='Cart_Completion_Frequency', data=data)
plt.title('Age vs Cart Completion Frequency')
plt.show()
```

4.2. Quan Hệ Giữa Biến Định Lượng và Biến Định Tính
Tần suất mua hàng dựa trên giới tính: Sử dụng box plot để kiểm tra phân phối của biến định lượng theo các nhóm của biến định tính.

```python
sns.boxplot(x='Gender', y='age', data=data)
plt.title('Age Distribution by Gender')
plt.show()
```

4.3. Quan Hệ Giữa Các Biến Định Tính
Bảng chéo (Crosstab) và biểu đồ cột: Sử dụng để kiểm tra quan hệ giữa các biến định tính.

```python
pd.crosstab(data['Gender'], data['Purchase_Frequency']).plot(kind='bar')
plt.title('Purchase Frequency by Gender')
plt.show()
```

5. Trực Quan Hóa Dữ Liệu
Histogram và KDE: Để hiểu phân phối của các biến.

```python
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

Box Plot: Để phát hiện các giá trị ngoại lai.

```python
sns.boxplot(x=data['age'])
plt.title('Age Box Plot')
plt.show()
```

Pair Plot: Để kiểm tra quan hệ giữa các cặp biến.

```python
sns.pairplot(data)
plt.show()
```

Heatmap: Để kiểm tra ma trận tương quan giữa các biến.
